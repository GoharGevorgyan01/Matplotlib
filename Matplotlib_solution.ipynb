ùê∏=ùëöùëê2

–° –ø–æ–º–æ—â—å—é –í–∞—à–∏—Ö –∑–Ω–∞–Ω–∏–π Numpy —Å–æ–∑–¥–∞–π—Ç–µ –¥–≤–∞ –º–∞—Å—Å–∏–≤–∞: E –∏ m:

    –∑–Ω–∞—á–µ–Ω–∏—è m - —ç—Ç–æ 11 —Ä–∞–≤–Ω–æ–æ—Ç—Å—Ç–æ—è—â–∏—Ö –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ —á–∏—Å–µ–ª –æ—Ç 0 –≥—Ä–∞–º–º–æ–≤ –¥–æ 10 –≥—Ä–∞–º–º–æ–≤
    –∑–Ω–∞—á–µ–Ω–∏—è E –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Ä–∞–≤–Ω—ã —ç–Ω–µ—Ä–≥–∏–∏ –¥–ª—è —ç—Ç–∏—Ö –º–∞—Å—Å.

–í–∞–º –Ω—É–∂–Ω–æ –±—É–¥–µ—Ç —É–∫–∞–∑–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è c, –ø—Ä–∏–≤–µ–¥—è –µ–≥–æ –≤ –Ω—É–∂–Ω—ã–µ –µ–¥–∏–Ω–∏—Ü—ã –∏–∑–º–µ—Ä–µ–Ω–∏—è - –≤ –º–µ—Ç—Ä–∞—Ö –≤ —Å–µ–∫—É–Ω–¥—É. –ú–æ–∂–µ—Ç–µ –Ω–∞–π—Ç–∏ –æ—Ç–≤–µ—Ç —Å –ø–æ–º–æ—â—å—é –ø–æ–∏—Å–∫–∞ –≤ Google (–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±—É–¥–µ—Ç —É–∫–∞–∑–∞—Ç—å –ø—Ä–∏–º–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ).

–ó–ê–ú–ï–ß–ê–ù–ò–ï: –ï—Å–ª–∏ –í–∞–º –Ω–µ–ø–æ–Ω—è—Ç–Ω–∞ –ø–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏, —Ç–æ –Ω–µ —Å—Ç–µ—Å–Ω—è–π—Ç–µ—Å—å –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å —Ä–µ—à–µ–Ω–∏–µ –≤ –±–ª–æ–∫–Ω–æ—Ç–µ —Å —Ä–µ—à–µ–Ω–∏—è–º–∏.

```python
!pip install numpy
import numpy as np
```

```python
m = np.linspace(0,10,11)
m
```

```python
E = np.round(m * ((3*10**8)**2),2)
E
```

### –ó–∞–¥–∞–Ω–∏–µ 1, —á–∞—Å—Ç—å 2: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫ E=mc^2

–¢–µ–ø–µ—Ä—å, –∫–æ–≥–¥–∞ —É –Ω–∞—Å –µ—Å—Ç—å –º–∞—Å—Å–∏–≤—ã E –∏ m, –º—ã –º–æ–∂–µ–º –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —ç–Ω–µ—Ä–≥–∏–∏ –æ—Ç –º–∞—Å—Å—ã. 

**–ó–∞–¥–∞–Ω–∏–µ: –≤—ã–ø–æ–ª–Ω–∏—Ç–µ –∫–æ–º–∞–Ω–¥—É Import –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ Matplotlib:**

```python
import matplotlib.pyplot as plt
```

**–ó–∞–¥–∞–Ω–∏–µ: –≤–æ—Å–ø—Ä–æ–∏–∑–≤–µ–¥–∏—Ç–µ –ø–æ–∫–∞–∑–∞–Ω–Ω—ã–π –Ω–∏–∂–µ –≥—Ä–∞—Ñ–∏–∫, –∏—Å–ø–æ–ª—å–∑—É—è —Ä–∞–Ω–µ–µ —Å–æ–∑–¥–∞–Ω–Ω—ã–µ –º–∞—Å—Å–∏–≤—ã. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Ü–≤–µ—Ç–∞, –æ—Å–∏ –∏ –Ω–∞–¥–ø–∏—Å–∏. –ù–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ —Å–æ–∑–¥–∞–≤–∞—Ç—å —Ç–æ—á–Ω–æ —Ç–∞–∫–æ–π –∂–µ –≥—Ä–∞—Ñ–∏–∫, –Ω–æ –ø–æ—Å—Ç–∞—Ä–∞–π—Ç–µ—Å—å –≤–æ—Å–ø—Ä–æ–∏–∑–≤–µ—Å—Ç–∏ –≤—Å–µ –æ—Å–Ω–æ–≤–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –≥—Ä–∞—Ñ–∏–∫–∞.**

```python
# –ü–ò–®–ò–¢–ï –í–ê–® –ö–û–î –ó–î–ï–°–¨
fig = plt.figure(figsize=(3,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(m,E,color = 'r',lw = 3)
axes.set_xlim(0,10)
axes.set_xlabel('Mass in Grams')
axes.set_ylabel('Energy in Joules')
axes.set_title('E=mc^2');
```

```python
# –ù–ï –ó–ê–ü–£–°–ö–ê–ô–¢–ï –Ø–ß–ï–ô–ö–£ –ù–ò–ñ–ï, –ò–ù–ê–ß–ï –ì–†–ê–§–ò–ö –ò–°–ß–ï–ó–ù–ï–¢!
```

### –ó–∞–¥–∞–Ω–∏–µ 1, —á–∞—Å—Ç—å 3 (–±–æ–Ω—É—Å - –æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ)

**–ú–æ–∂–µ—Ç–µ –ª–∏ –í—ã –≤—ã—è—Å–Ω–∏—Ç—å, –∫–∞–∫ –ø—Ä–∏ —Ä–∏—Å–æ–≤–∞–Ω–∏–∏ —ç—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ –ø–æ–ª—É—á–∏—Ç—å –Ω–∞ –æ—Å–∏ y –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫—É—é —à–∫–∞–ª—É? –¢–∞–∫–∂–µ –Ω–∞–Ω–µ—Å–∏—Ç–µ —Å–µ—Ç–∫—É (grid) –ø–æ –æ—Å–∏ y. –ú—ã –Ω–µ —Ä–∞–∑–±–∏—Ä–∞–ª–∏ —ç—Ç–∏ –≤–µ—â–∏ –≤ –≤–∏–¥–µ–æ-–ª–µ–∫—Ü–∏—è—Ö, —Ç–∞–∫ —á—Ç–æ –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –Ω–∞–π—Ç–∏ –æ—Ç–≤–µ—Ç—ã –ø–æ–∏—Å–∫–æ–º –≤ Google, StackOverflow, –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏ –ø–æ Matplotlib, –∏–ª–∏ –¥–∞–∂–µ –≤ –Ω–∞—à–µ–º –±–ª–æ–∫–Ω–æ—Ç–µ "Additional Matplotlib Commands". –í —ç—Ç–æ–º –≥—Ä–∞—Ñ–∏–∫–µ –Ω—É–∂–Ω–æ –¥–æ–±–∞–≤–∏—Ç—å –µ—â—ë —Ç–æ–ª—å–∫–æ –¥–≤–µ —Å—Ç—Ä–æ–∫–∏ –∫–æ–¥–∞.**

```python
# –ü–ò–®–ò–¢–ï –í–ê–® –ö–û–î –ó–î–ï–°–¨
fig = plt.figure(figsize=(3,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(m,E,color = 'r',lw = 3)
axes.set_xlim(0,10)
axes.set_xlabel('Mass in Grams')
axes.set_ylabel('Energy in Joules')
axes.set_title('E=mc^2')
plt.yscale("log")
axes.grid(axis = 'y',which = 'both');
```

## –ó–∞–¥–∞–Ω–∏–µ 2, —á–∞—Å—Ç—å 1: –î–∞–Ω–Ω—ã–µ –¥–ª—è –∫—Ä–∏–≤–æ–π –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏

**–ú—ã –≤–∑—è–ª–∏ –¥–ª—è –í–∞—Å –Ω—É–∂–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∫—Ä–∏–≤–æ–π –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏ —Å —Å–∞–π—Ç–∞ [US Treasury Dept.](https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx?data=yield) –î–∞–Ω–Ω—ã–µ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç –ø—Ä–æ—Ü–µ–Ω—Ç–Ω—ã–µ —Å—Ç–∞–≤–∫–∏ –æ–±–ª–∏–≥–∞—Ü–∏–π –°–®–ê –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Å—Ä–æ–∫–æ–≤ –∫–æ–Ω—Ç—Ä–∞–∫—Ç–∞. –°–ø–∏—Å–æ–∫ labels –æ—Ç–æ–±—Ä–∞–∂–∞–µ—Ç —Å—Ä–æ–∫–∏ –∫–æ–Ω—Ç—Ä–∞–∫—Ç–∞ –¥–ª—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –ø–æ–∑–∏—Ü–∏–π –≤ —Å–ø–∏—Å–∫–∞—Ö –∑–Ω–∞—á–µ–Ω–∏–π.**

**–ó–∞–¥–∞–Ω–∏–µ: –ó–∞–ø—É—Å—Ç–∏—Ç–µ —è—á–µ–π–∫—É –Ω–∏–∂–µ, —á—Ç–æ–±—ã —Å–æ–∑–¥–∞—Ç—å —Å–ø–∏—Å–∫–∏ –¥–ª—è –ø–æ—Å–ª–µ–¥—É—é—â–µ–≥–æ —Ä–∏—Å–æ–≤–∞–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞.**

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

**–ó–∞–¥–∞–Ω–∏–µ: –≤—ã—è—Å–Ω–∏—Ç–µ, –∫–∞–∫ –º–æ–∂–Ω–æ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –æ–±–µ —ç—Ç–∏ –ª–∏–Ω–∏–∏ –Ω–∞ –æ–¥–Ω–æ–º –∏ —Ç–æ–º –∂–µ –≥—Ä–∞—Ñ–∏–∫–µ. –î–æ–±–∞–≤—å—Ç–µ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –ª–µ–≥–µ–Ω–¥—É.**

```python
# –ù–ï –ó–ê–ü–£–°–ö–ê–ô–¢–ï –Ø–ß–ï–ô–ö–£ –ù–ò–ñ–ï, –ò–ù–ê–ß–ï –ì–†–ê–§–ò–ö –ò–°–ß–ï–ó–ù–ï–¢!
```

**–ó–∞–¥–∞–Ω–∏–µ: —Ö–æ—Ç—è –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ —è—Å–Ω–æ –≤–∏–¥–Ω–æ —Ä–∞–∑–Ω–∏—Ü—É –º–µ–∂–¥—É 2007 –∏ 2020 –≥–æ–¥–∞–º–∏, –∫–æ–≥–¥–∞ –º—ã –Ω–∞–Ω–æ—Å–∏–º —ç—Ç–∏ –ª–∏–Ω–∏–∏ –Ω–∞ –æ–¥–∏–Ω –∏ —Ç–æ—Ç –∂–µ –≥—Ä–∞—Ñ–∏–∫, –Ω–∞–º —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —Å–ª–æ–∂–Ω–µ–µ —É–≤–∏–¥–µ—Ç—å –¥–µ—Ç–∞–ª–∏ –≤–Ω—É—Ç—Ä–∏ –∫–∞–∂–¥–æ–≥–æ –æ—Ç–¥–µ–ª—å–Ω–æ–≥–æ –≥–æ–¥–∞. –° –ø–æ–º–æ—â—å—é .suplots() –Ω–∞—Ä–∏—Å—É–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫–∏ –Ω–∏–∂–µ, –≤ –∫–æ—Ç–æ—Ä—ã—Ö —ç—Ç–∏ –¥–µ—Ç–∞–ª–∏ –±—É–¥—É—Ç –ª—É—á—à–µ –≤–∏–¥–Ω—ã.**

```python
# –ü–ò–®–ò–¢–ï –í–ê–® –ö–û–î –ó–î–ï–°–¨
fig, axes = plt.subplots(nrows=2,ncols=1,figsize = (7,5))
axes[0].plot(labels,july16_2007)
axes[1].plot(labels,july16_2020)
axes[0].set_title('July 16th,2007')
axes[1].set_title('July 16th,2020')
plt.tight_layout();
```

```python
# –ù–ï –ó–ê–ü–£–°–ö–ê–ô–¢–ï –Ø–ß–ï–ô–ö–£ –ù–ò–ñ–ï, –ò–ù–ê–ß–ï –ì–†–ê–§–ò–ö –ò–°–ß–ï–ó–ù–ï–¢!
```

**–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –±–æ–Ω—É—Å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ: –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –≤–æ—Å–ø—Ä–æ–∏–∑–≤–µ—Å—Ç–∏ –≥—Ä–∞—Ñ–∏–∫ –Ω–∏–∂–µ, —Å –¥–≤—É–º—è –æ—Å—è–º–∏. –•–æ—Ç—è –≤ –¥–∞–Ω–Ω–æ–º –ø—Ä–∏–º–µ—Ä–µ –¥–≤–µ —Ä–∞–∑–Ω—ã–µ –æ—Å–∏ –ø–æ —Å—É—Ç–∏ –Ω–µ –Ω—É–∂–Ω—ã, —ç—Ç–æ —Ö–æ—Ä–æ—à–µ–µ —É–ø—Ä–∞–∂–Ω–µ–Ω–∏–µ –ø–æ Matplotlib, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–∏–≥–æ–¥–∏—Ç—Å—è –≤ –¥—Ä—É–≥–∏—Ö —Å–∏—Ç—É–∞—Ü–∏—è—Ö.**

```python
# –ü–ò–®–ò–¢–ï –í–ê–® –ö–û–î –ó–î–ï–°–¨
fig, ax1 = plt.subplots()
ax1.spines['left'].set_color('blue')
ax1.spines['left'].set_linewidth(3)


ax1.plot(labels,july16_2007, lw=2, color="blue")
ax1.set_ylabel("2007", fontsize=18, color="blue")
ax1.set_title('July 16th Yield Curves')
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(labels,july16_2020, lw=2, color="red")
ax2.set_ylabel("2020", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")

ax2.spines['right'].set_color('red')
ax2.spines['right'].set_linewidth(3)
```
